```python
import matplotlib.pyplot as plt
import struct
```

```python
# Function to read data from a file
def read_data_txt(filename):
    x_vals, y_vals = [], []
    with open(filename, 'r') as file:
        for line in file:
            try:
                x, y = map(float, line.split())
                x_vals.append(x)
                y_vals.append(y)
            except ValueError:
                print(f"Skipping invalid line: {line.strip()}")
    return x_vals, y_vals

def read_resrel_bin(filename):
    res_vals, rel_vals = [], []
    record_size = 8

    with open(filename, 'rb') as file:
        while (True):

            record = file.read(record_size)
            if len(record) < record_size:
                break
            res, rel = struct.unpack("ff", record)
            res.append(res)
            rel.append(rel)
    return res_vals, rel_vals


```

```python
x, y = read_data_txt("./partitions/partitions_of_64_resrel.txt")
res_vals, rel_vals = read_resrel_bin ("./partitions/partitions_of_64_resrel.bin")
```

```python
# Function to plot scatter plot
plt.scatter(res_vals, y, color='blue', alpha=0.7, marker=".", linewidths=0.7)
plt.xlabel('Resolution')
plt.ylabel('Relevance')
plt.title('n = 64')
plt.grid(True)
plt.show()
```
